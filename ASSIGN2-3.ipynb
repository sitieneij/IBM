```python
df['Education'].value_counts().plot(kind = 'bar')
df['DistanceFromHome'].value_counts().plot(kind ='bar')
```

```python
import matplotlib.pyplot as plt

keys = [Age for Age, df in df1.groupby(['Age'])]

plt.bar(keys, df.groupby(['Age']).sum()['Number']
plt.ylabel('Age')
plt.xlabel('Number')
plt.xticks(keys, rotation='vertical', size=8)
plt.show()
```
