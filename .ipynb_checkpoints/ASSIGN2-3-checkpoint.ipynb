```python
#import matplotlib.style as mplstyle
#mplstyle.use('fast')
import numpy as np
x = np.linspace(1, 10, 20)
y = np.linspace(20, 30, 40)
# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='YearsAtCompany')  # Plot some data on the axes.
ax.plot(x, x**2, label='YearsAtCurrentRole')  # Plot more data on the axes...
ax.plot(x, x**3, label='YearsSinceLastPromotion')  # ... and some more.
ax.set_xlabel('Years')  # Add an x-label to the axes.
ax.set_ylabel('Age')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```
